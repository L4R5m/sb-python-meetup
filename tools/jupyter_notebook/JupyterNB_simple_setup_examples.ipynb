## Util Fns

````python
# Config mpl rcParams
def configMpl(interactive=False, restoreDefaults=False, customizePlt=True):
    """Configures pyplot to plot inline (static) or with 'notebook'
    backend (interactive).
    :param interactive: bool, if True mpl creates interactive plots
    :param restoreDefaults: bool, if True restores default rcParams
    :param customizePlt: bool, if True customizes plt rcParams to be pretty
    """
    if interactive:
        %matplotlib notebook
    else:
        %matplotlib inline
        
    # Need to re-import pyplot when changing backend
    import matplotlib.pyplot as plt
    if restoreDefaults:
        plt.rcdefaults()
    elif customizePlt:
        tikClr = '#EEEEEE'
        figClr = (0.2, 0.2, 0.2)
        figTitleSize = 20
        plt.rcParams['figure.figsize'] = (10,8)
        plt.rcParams['xtick.color'] = tikClr
        plt.rcParams['ytick.color'] = tikClr
        plt.rcParams['figure.facecolor'] = figClr
        plt.rcParams['axes.facecolor'] = figClr
        plt.rcParams['figure.titlesize'] = figTitleSize
        plt.rcParams['grid.alpha'] = 0.3

configMpl(True, False)
````

# Data Wrangling & Exploration

````python
import pandas
import pathlib as pl
````

## Get data and cache to local file

````python
# Define source url
# url, dataType = "https://data.smcgov.org/resource/mb6a-xn89.json", 'json'
url, dataType = "https://raw.githubusercontent.com/plotly/datasets/master/tips.csv", 'csv'
````

````python
import requests
resp = requests.get(url)
resp
````

````python
resp.content
````

````python
fpth = pl.Path('.', f'data.{dataType}')
with open(fpth, 'wb') as fh:
    fh.write(resp.content)
````

## Load data into DataFrame

````python
# Show fn docstr
pd.set_option?
````

````python
# Set pd display options
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 180)
pd.set_option('display.max_colwidth', 25)
````

````python
# Load data
print(f'Loading {fpth} into DataFrame')
if dataType == 'json':
    df = pd.read_json(fpth)
elif dataType == 'csv':
    df = pd.read_csv(fpth)
else:
    raise NotImplementedError(f'Unsupported data type: {dataType}')

df
````

````python
# Add derived column
df['pct'] = df.tip / df.total_bill
df
````

````python
df.shape, df.describe()
````

````python
# Handle various source datas
if dataType == 'json':
    df.plot("less_than_high_school_graduate", "high_school_graduate", ls='', marker='o')
elif dataType == 'csv':
    df.plot(kind='scatter', x='total_bill', y='pct', title='Pct Tip vs Total Bill')
````
